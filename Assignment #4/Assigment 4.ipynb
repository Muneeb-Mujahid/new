```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("netflix_data.csv")
data.head()
```

```python
data.isnull().sum()
```

### Q1: In the netflix dataset replace all null values present in country column with united states.

```python
data["country"].replace(np.nan, "United States",inplace=True)
```

```python
data.head(n=100)
```

```python
data.isnull().sum()
```

# Q2: Show 10 countries with most tv-shows and movies

```python
country = data.groupby("type")["country"].value_counts()
print(country['Movie'].head(10))
print(country['TV Show'].head(10))
```

# Q3: Count of movies released in year respective years. example

```python
Count_of_movies = data.groupby("type")["release_year"].value_counts()
print(Count_of_movies["Movie"])
```

# Q4: Similarly, count of tv-shows released in there respective years.

```python
Count_of_tv_shows = data.groupby("type")["release_year"].value_counts()
print(Count_of_tv_shows["TV Show"])
```
